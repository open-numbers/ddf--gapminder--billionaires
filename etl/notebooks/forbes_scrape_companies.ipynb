```python
def get_text(person):
    p = person.replace('_', '-')
    url = profile_tmpl.format(p)
    res = req.get(url)
    if not res.ok:
        print(f'{person}: http problem: {res.status_code}')
        return (person, None)
    root = lxml.html.fromstring(res.content)
    e = root.xpath("//div[@class='profile-text']")
    if len(e) != 1:
        print(f'{person}: html problem')
        return (person, None)
    text = '\n'.join(e[0].itertext())
    return (person, text)
```

```python
%time get_text('zong_yanmin')
```

```python
with multiprocessing.Pool(2) as pool:
    res = pool.map(get_text, all_forbes.person.values[:3])
```
