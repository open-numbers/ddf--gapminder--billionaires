```python
def age_today(t):
    bd = datetime.strptime(t, '%Y-%m-%d')
    td = datetime.today()
    
    try:
        currentYrAnniversary = datetime( td.year, bd.month, bd.day )
    except ValueError:
        currentYrAnniversary = datetime( td.year, bd.month, bd.day-1 )
    
    return (td.year - bd.year) - (1 if td < currentYrAnniversary else 0)
```

```python
def _f(x):
    if pd.isnull(x):
        return x
    return age_today(x)

df1['age'] = df1['hs_Character_Birthday'].map(_f)
```

```python
def name_to_id(name):
    if not isinstance(name, str):
        return name
    n = name.lower().strip()
    if '& family' in n:
        n = to_concept_id(n.replace('& family', ''))
    elif '& famiy' in n:
        n = to_concept_id(n.replace('& famiy', ''))
    elif '&family' in n:
        n = to_concept_id(n.replace('&family', ''))
    elif ' and family' in n:
        n = to_concept_id(n.replace(' and family', ''))
    elif ' brothers' in n:
        n = to_concept_id(n.replace(' brothers', ''))
#     elif 'family' in n:
#         n = to_concept_id(n.replace('family', ''))
    elif '&' in n:
        n = to_concept_id(n.replace('&', ' and '))
    elif '、' in n:
        n = to_concept_id(n.replace('、', ' and '))
    else:
        n = to_concept_id(n)
        
    return n
```

```python
def get_country(x):
    if x == '':
        return None
    if 'Taipei' in x:
        return "Taiwan"
    if 'Hong Kong' in x:
        return "Hong Kong"
    if x == 'ChinaChangsha':
        return 'China'
    if x == 'United StatesNew Jersey':
        return "United States"
    res = x.split('-')[0].strip()
    if res == 'Dubai':
        return "United Arab Emirates"
    if res == 'Melbourne':
        return 'Australia'
    return res

def run_get_country(x):
    if ';' in x:
        splits = x.split(';')
        return ';'.join(set(map(get_country, splits)))
    else:
        return get_country(x)

res_df2['country'] = res_df2['permanent_country'].map(run_get_country)
```

```python
# fix age issue in birth_2
def fix_age(x):
    if x == '':
        return np.nan
    if pd.isnull(x):
        return x
    if x == '/' or x == '#VALUE!':
        return np.nan

    if '、' in x:
        x = x.replace('、', ';')
    if ';' in x:
        split = x.split(';')
        res = filter(lambda x: x not in ['#VALUE!', '/'], split)
        return ';'.join(res)
    return x
```
